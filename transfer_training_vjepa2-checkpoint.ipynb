```python
import os
import torch
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.io as io
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV, learning_curve, validation_curve, StratifiedKFold
import numpy as np
from transformers import AutoModel, AutoVideoProcessor
import matplotlib.pyplot as plt
import seaborn as sns
import random
import joblib # Â∞éÂÖ• joblib Áî®ÊñºÂÑ≤Â≠òÂíåËºâÂÖ•Ê®°Âûã

# Ë®≠ÂÆöÈö®Ê©üÁ®ÆÂ≠ê‰ª•Á¢∫‰øùÁµêÊûúÂèØÈáçÁèæÊÄß
np.random.seed(42)
torch.manual_seed(42)
random.seed(42)

VIDEO_DIR = "videos"
LABELS = sorted(os.listdir(VIDEO_DIR)) # ÂÅáË®≠ÈÄôË£°ÊúÉËÆÄÂèñ‰Ω†ÁöÑÂΩ±ÁâáË≥áÊñôÂ§æÔºåÁç≤ÂèñÊâÄÊúâÈ°ûÂà•
FRAMES_PER_CLIP = 16
BATCH_SIZE = 1 # ÁâπÂæµÊèêÂèñÈöéÊÆµ‰øùÊåÅ batch size ÁÇ∫ 1
NUM_WORKERS = 4
HF_REPO = "facebook/vjepa2-vitg-fpc64-256"

label2id = {label: idx for idx, label in enumerate(LABELS)}
id2label = {idx: label for idx, label in enumerate(LABELS)} # Á¢∫‰øù id2label ‰πüÊ≠£Á¢∫ÂàùÂßãÂåñ

# --- KTHVideoDataset È°ûÂà• (‰∏çËÆä) ---
class KTHVideoDataset(Dataset):
    def __init__(self, video_dir, processor, frames_per_clip=16, max_per_class=100):
        self.samples = []
        self.processor = processor
        self.frames_per_clip = frames_per_clip

        for label in os.listdir(video_dir):
            class_dir = os.path.join(video_dir, label)
            if not os.path.isdir(class_dir): continue
            video_files = [f for f in os.listdir(class_dir) if f.endswith(".avi")]
            selected = random.sample(video_files, min(len(video_files), max_per_class))
            for fname in selected:
                path = os.path.join(class_dir, fname)
                self.samples.append((path, label2id[label]))

        print(f"üé† Found {len(self.samples)} video samples.")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        video, _, _ = io.read_video(path, pts_unit='sec')
        total_frames = video.shape[0]
        if total_frames >= self.frames_per_clip:
            start = (total_frames - self.frames_per_clip) // 2
            clip = video[start:start+self.frames_per_clip]
        else:
            pad = self.frames_per_clip - total_frames
            clip = torch.cat([video, video[-1:].repeat(pad, 1, 1, 1)])
        
        clip_list = [frame.numpy() for frame in clip]
        inputs = self.processor(clip_list, return_tensors="pt")
        inputs = {k: v.squeeze(0) for k, v in inputs.items()}
        return inputs, label

# --- extract_features ÂáΩÊï∏ (‰∏çËÆä) ---
def extract_features(base_model, dataset):
    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS)
    pool = torch.nn.AdaptiveAvgPool1d(1)
    features, labels = [], []
    base_model.cuda().eval()

    print(f"Extracting features from {len(dataset)} samples...")
    with torch.no_grad():
        for i, (x, y) in enumerate(loader):
            x = {k: v.cuda(non_blocking=True) for k, v in x.items()}
            output = base_model(**x)
            last_hidden = output.last_hidden_state
            pooled = pool(last_hidden.permute(0, 2, 1)).squeeze(-1)
            features.append(pooled.cpu().numpy()[0])
            labels.append(y.item())
            
            if (i + 1) % 100 == 0:
                print(f"Processed {i+1}/{len(dataset)} samples.")

    features = np.array(features)
    labels = np.array(labels)
    print("‚úÖ Feature extraction complete.")
    return features, labels

# --- plot_learning_curve ÂáΩÊï∏ (‰∏çËÆä) ---
def plot_learning_curve(estimator, title, X, y, axes=None, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    if axes is None:
        _, axes = plt.subplots(1, 1, figsize=(10, 6))

    axes.set_title(title)
    if ylim is not None:
        axes.set_ylim(*ylim)
    axes.set_xlabel("Training examples")
    axes.set_ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = \
        learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                       train_sizes=train_sizes,
                       return_times=True, scoring='accuracy')
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    # fit_times_mean = np.mean(fit_times, axis=1) # ÁßªÈô§‰∏çÂøÖË¶ÅÁöÑËÆäÊï∏
    # fit_times_std = np.std(fit_times, axis=1) # ÁßªÈô§‰∏çÂøÖË¶ÅÁöÑËÆäÊï∏

    axes.grid()
    axes.fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    axes.fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
    axes.plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
    axes.plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
    axes.legend(loc="best")
    return plt

# --- plot_validation_curve ÂáΩÊï∏ (‰∏çËÆä) ---
def plot_validation_curve(estimator, title, X, y, param_name, param_range, axes=None, ylim=None, cv=None,
                          n_jobs=None, scoring='accuracy'):
    if axes is None:
        _, axes = plt.subplots(1, 1, figsize=(10, 6))

    axes.set_title(title)
    if ylim is not None:
        axes.set_ylim(*ylim)
    axes.set_xlabel(param_name)
    axes.set_ylabel("Score")

    train_scores, test_scores = validation_curve(
        estimator, X, y, param_name=param_name, param_range=param_range,
        cv=cv, n_jobs=n_jobs, scoring=scoring
    )
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    axes.grid()
    axes.plot(param_range, train_scores_mean, label="Training score",
                 color="darkorange", lw=2)
    axes.fill_between(param_range, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.2,
                         color="darkorange", lw=2)
    axes.plot(param_range, test_scores_mean, label="Cross-validation score",
                 color="navy", lw=2)
    axes.fill_between(param_range, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.2,
                         color="navy", lw=2)
    axes.legend(loc="best")
    return plt

# --- train_and_evaluate_model ÂáΩÊï∏ (Âè™ÁßªÈô§ÂñÆÁç®ÂÑ≤Â≠òÂêÑÊ®°ÂûãÁöÑÈÇèËºØ) ---
def train_and_evaluate_model(model_name, estimator, param_grid, train_features, train_labels, val_features, val_labels, labels_map, cv_strategy):
    """
    Trains and evaluates a given model, performs cross-validation, plots learning/validation curves,
    and returns the best estimator for ROC curve comparison.
    """
    pipeline = make_pipeline(StandardScaler(), estimator)
    print(f"\n--- Training {model_name} ---")

    # GridSearchCV for hyperparameter tuning with cross-validation
    clf = GridSearchCV(pipeline, param_grid, cv=cv_strategy, n_jobs=-1, verbose=1, scoring='accuracy')
    clf.fit(train_features, train_labels)

    print(f"\nüèÜ Best parameters for {model_name}: {clf.best_params_}")
    print(f"Cross-validation mean score for {model_name}: {clf.best_score_:.4f}")

    best_model = clf.best_estimator_
    
    # ÈÄôË£°‰∏çÂÜçÂñÆÁç®ÂÑ≤Â≠òÊØèÂÄãÊ®°ÂûãÁöÑÊúÄ‰Ω≥ÁâàÊú¨ÔºåÂõ†ÁÇ∫ÊàëÂÄëÁõÆÊ®ôÊòØÂè™ÂÑ≤Â≠òÁ∏ΩÈ´îÊúÄ‰Ω≥ÁöÑÊ®°Âûã
    # joblib.dump(best_model, f"vjepa_classifier_{model_name.replace(' ', '_').lower()}.pt")
    # print(f"Model saved: vjepa_classifier_{model_name.replace(' ', '_').lower()}.pt")


    # Evaluation on the validation set
    print(f"\nüß™ Evaluating {model_name} on the validation set...")
    val_preds = best_model.predict(val_features)
    print(f"\n--- Validation Set Performance Report for {model_name} ---")
    print(classification_report(val_labels, val_preds, target_names=labels_map))

    # Plot confusion matrix
    cm = confusion_matrix(val_labels, val_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=labels_map, yticklabels=labels_map, cmap="Blues")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Validation Set Confusion Matrix for {model_name}")
    plt.tight_layout()
    plt.show()

    # Plot Learning Curve
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    plot_learning_curve(pipeline, f"Learning Curve ({model_name})", train_features, train_labels,
                        axes=ax, cv=cv_strategy, n_jobs=-1)
    plt.show()

    # Plot Validation Curve (example for one hyperparameter, needs adjustment for others)
    if model_name == "Random Forest":
        param_name = "randomforestclassifier__n_estimators"
        param_range = [50, 100, 200, 300]
    elif model_name == "SVM":
        param_name = "svc__C"
        param_range = np.logspace(-1, 2, 4)
    elif model_name == "XGBoost":
        param_name = "xgbclassifier__n_estimators"
        param_range = [50, 100, 200, 300]
    else:
        param_name = None

    if param_name:
        fig, ax = plt.subplots(1, 1, figsize=(10, 6))
        plot_validation_curve(pipeline, f"Validation Curve ({model_name}) - {param_name.split('__')[-1]}",
                              train_features, train_labels, param_name=param_name, param_range=param_range,
                              axes=ax, cv=cv_strategy, n_jobs=-1)
        plt.show()

    return best_model # Return the best trained model for ROC comparison

# --- plot_roc_curves ÂáΩÊï∏ (‰∏çËÆä) ---
def plot_roc_curves(models, val_features, val_labels, labels_map):
    """
    Plots ROC curves for multiple models for each class.
    Assumes models can output probability estimates (predict_proba).
    """
    plt.figure(figsize=(12, 10))
    # Ë®àÁÆóÂ≠êÂúñË°åÊï∏ÂíåÂàóÊï∏ÔºåÁ¢∫‰øù‰ΩàÂ±ÄÂêàÁêÜ
    num_classes = len(labels_map)
    rows = (num_classes + 1) // 2 # Á¢∫‰øùËá≥Â∞ë‰∏ÄË°å
    cols = 2 if num_classes > 1 else 1 # Â¶ÇÊûúÂè™Êúâ‰∏ÄÈ°ûÂâá‰∏ÄÂàó

    for i, label_name in enumerate(labels_map):
        plt.subplot(rows, cols, i + 1)
        for model_name, model in models.items():
            if hasattr(model, "predict_proba"):
                y_score = model.predict_proba(val_features)[:, i]
                fpr, tpr, _ = roc_curve(val_labels, y_score, pos_label=i)
                roc_auc = auc(fpr, tpr)
                plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
            else:
                print(f"Warning: Model {model_name} does not support predict_proba, skipping ROC curve for class {label_name}.")

        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for class: {label_name}')
        plt.legend(loc="lower right")
    plt.tight_layout()
    plt.suptitle("ROC Curve Comparison Across Models and Classes", y=1.02, fontsize=16)
    plt.show()


def main():
    processor = AutoVideoProcessor.from_pretrained(HF_REPO)
    base_model = AutoModel.from_pretrained(HF_REPO)
    full_dataset = KTHVideoDataset(VIDEO_DIR, processor, FRAMES_PER_CLIP, max_per_class=50)

    # 7:3 ÂàÜÂâ≤Ë≥áÊñôÈõÜ
    train_size = int(0.7 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    # ÁÇ∫Ë®ìÁ∑¥ÈõÜÂíåÈ©óË≠âÈõÜÂàÜÂà•ÊèêÂèñÁâπÂæµ
    print("üì¶ ËêÉÂèñ„ÄåË®ìÁ∑¥ÈõÜ„ÄçÁâπÂæµ‰∏≠...")
    train_features, train_labels = extract_features(base_model, train_dataset)

    print("üì¶ ËêÉÂèñ„ÄåÈ©óË≠âÈõÜ„ÄçÁâπÂæµ‰∏≠...")
    val_features, val_labels = extract_features(base_model, val_dataset)

    # ÂÆöÁæ©ÂÖ±ÂêåÁöÑ‰∫§ÂèâÈ©óË≠âÁ≠ñÁï•
    cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    trained_models = {} # ÂÑ≤Â≠òÊØèÂÄãÊ®°ÂûãÈ°ûÂûãÂêÑËá™ÁöÑÊúÄ‰Ω≥Ê®°ÂûãÁâ©‰ª∂
    validation_accuracies = {} # ÂÑ≤Â≠òÊØèÂÄãÊ®°ÂûãÂú®È©óË≠âÈõÜ‰∏äÁöÑÊ∫ñÁ¢∫Â∫¶

    # --- Random Forest ---
    rf_param_grid = {
        'randomforestclassifier__n_estimators': [100, 200, 300],
        'randomforestclassifier__max_depth': [None, 10, 20],
    }
    rf_best_model = train_and_evaluate_model("Random Forest", RandomForestClassifier(random_state=42),
                                             rf_param_grid, train_features, train_labels, val_features, val_labels,
                                             LABELS, cv_strategy)
    trained_models["Random Forest"] = rf_best_model
    # Ë®àÁÆó‰∏¶ÂÑ≤Â≠ò RF Âú®È©óË≠âÈõÜ‰∏äÁöÑÊ∫ñÁ¢∫Â∫¶
    rf_val_preds = rf_best_model.predict(val_features)
    rf_accuracy = np.mean(rf_val_preds == val_labels)
    validation_accuracies["Random Forest"] = rf_accuracy
    print(f"Random Forest Validation Accuracy: {rf_accuracy:.4f}")

    # --- SVM ---
    svm_param_grid = {
        'svc__C': [0.1, 1, 10],
        'svc__gamma': ['scale', 'auto'],
        'svc__kernel': ['rbf']
    }
    svm_best_model = train_and_evaluate_model("SVM", SVC(random_state=42, probability=True), # probability=True ÂïüÁî® predict_proba
                                              svm_param_grid, train_features, train_labels, val_features, val_labels,
                                              LABELS, cv_strategy)
    trained_models["SVM"] = svm_best_model
    # Ë®àÁÆó‰∏¶ÂÑ≤Â≠ò SVM Âú®È©óË≠âÈõÜ‰∏äÁöÑÊ∫ñÁ¢∫Â∫¶
    svm_val_preds = svm_best_model.predict(val_features)
    svm_accuracy = np.mean(svm_val_preds == val_labels)
    validation_accuracies["SVM"] = svm_accuracy
    print(f"SVM Validation Accuracy: {svm_accuracy:.4f}")

    # --- XGBoost ---
    xgb_param_grid = {
        'xgbclassifier__n_estimators': [100, 200, 300],
        'xgbclassifier__max_depth': [3, 5, 7],
        'xgbclassifier__learning_rate': [0.1, 0.05]
    }
    xgb_best_model = train_and_evaluate_model("XGBoost", XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'),
                                              xgb_param_grid, train_features, train_labels, val_features, val_labels,
                                              LABELS, cv_strategy)
    trained_models["XGBoost"] = xgb_best_model
    # Ë®àÁÆó‰∏¶ÂÑ≤Â≠ò XGBoost Âú®È©óË≠âÈõÜ‰∏äÁöÑÊ∫ñÁ¢∫Â∫¶
    xgb_val_preds = xgb_best_model.predict(val_features)
    xgb_accuracy = np.mean(xgb_val_preds == val_labels)
    validation_accuracies["XGBoost"] = xgb_accuracy
    print(f"XGBoost Validation Accuracy: {xgb_accuracy:.4f}")

    # Áπ™Ë£ΩÊâÄÊúâÊ®°ÂûãÁöÑ ROC Êõ≤Á∑ö
    print("\nüìä Áπ™Ë£Ω ROC Êõ≤Á∑ö‰ª•ÈÄ≤Ë°åÊØîËºÉ...")
    plot_roc_curves(trained_models, val_features, val_labels, LABELS)

    # --- ÊâæÂá∫‰∏¶ÂÑ≤Â≠òÊâÄÊúâÊ®°Âûã‰∏≠Âú®È©óË≠âÈõÜ‰∏äË°®ÁèæÊúÄÂ•ΩÁöÑÈÇ£‰∏ÄÂÄã ---
    print("\n--- ÊØîËºÉÊâÄÊúâÊ®°ÂûãÂú®È©óË≠âÈõÜ‰∏äÁöÑË°®Áèæ‰∏¶ÂÑ≤Â≠òÊúÄÁµÇÊúÄ‰Ω≥Ê®°Âûã ---")
    if validation_accuracies:
        # ÊâæÂà∞Ê∫ñÁ¢∫Â∫¶ÊúÄÈ´òÁöÑÊ®°ÂûãÂêçÁ®±
        overall_best_model_name = max(validation_accuracies, key=validation_accuracies.get)
        # Áç≤ÂèñË©≤Ê®°ÂûãÁâ©‰ª∂
        overall_best_model = trained_models[overall_best_model_name]
        overall_best_score = validation_accuracies[overall_best_model_name]

        # Â∞áÊï¥È´îÊúÄ‰Ω≥Ê®°ÂûãÂÑ≤Â≠òÂà∞Ê™îÊ°à
        joblib.dump(overall_best_model, "vjepa_classifier_overall_best.pt")
        print(f"\n‚ú® **Êï¥È´îÊúÄ‰Ω≥Ê®°ÂûãÁÇ∫: {overall_best_model_name}**")
        print(f"ÂÖ∂Âú®È©óË≠âÈõÜ‰∏äÁöÑÊ∫ñÁ¢∫Â∫¶ÁÇ∫: {overall_best_score:.4f}")
        print(f"Êï¥È´îÊúÄ‰Ω≥Ê®°ÂûãÂ∑≤ÂÑ≤Â≠òÁÇ∫: **vjepa_classifier_overall_best.pt**")
    else:
        print("Ê≤íÊúâÂèØÊØîËºÉÁöÑÊ®°ÂûãÊÄßËÉΩÊï∏Êìö„ÄÇ")

    # ÂÑ≤Â≠òÊ®ôÁ±§Â∞çÊáâÈóú‰øÇ
    with open("label_map.txt", "w") as f:
        for idx, label in id2label.items():
            f.write(f"{idx} {label}\n")
    print("‚úÖ Ë®ìÁ∑¥ÂÆåÊàêËàá label_map ÂÑ≤Â≠ò„ÄÇ")


if __name__ == "__main__":
    main()
```
